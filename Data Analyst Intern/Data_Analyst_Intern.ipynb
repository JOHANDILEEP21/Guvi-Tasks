## Analysing the Excel Dataset, sheet name= 'Sales Data'

```python
df1 = pd.read_excel('https://raw.githubusercontent.com/JOHANDILEEP21/Guvi-Tasks/main/Data%20Analyst%20Intern/Data%20Analyst%20Task.xls', sheet_name='Sales data')
df1
```

```python
df1.info()
```

```python
df1.isnull().sum()
```

```python
# Getting month year from order_date
df1['month_year'] = df1['Sales Date'].apply(lambda x: x.strftime('%Y-%m-%d'))
```

```python
# grouping month_year by sales
df_temp = df1.groupby('month_year').sum()['Product Amount with GST'].reset_index()
```

```python
# Setting the figure size
plt.figure(figsize=(16, 5))
plt.plot(df_temp['month_year'], df_temp['Product Amount with GST'], color='#b80045')
plt.xticks(rotation='vertical', size=8)
plt.show()
```

## WHICH ARE THE TOP 10 PRODUCTS BY SALES?


```python
# Grouping products by sales
prod_sales = pd.DataFrame(df1.groupby('Product Code').sum()['Product Amount with GST'])

# Sorting the dataframe in descending order
prod_sales.sort_values(by=['Product Amount with GST'], inplace=True, ascending=False)

# Top 10 products by sales
prod_sales[:10]
```

## WHICH ARE THE MOST SELLING PRODUCTS?

```python
# Grouping products by Quantity
best_selling_prods = pd.DataFrame(df1.groupby('Product Code').sum()['Product Amount with GST'])

# Sorting the dataframe in descending order
best_selling_prods.sort_values(by=['Product Amount with GST'], inplace=True, ascending=False)

# Most selling products
best_selling_prods[:10]
```

## WHAT IS THE MOST PREFERRED SHIP MODE?

```python
# Setting the figure size
plt.figure(figsize=(10, 8))

# countplot: Show the counts of observations in each categorical bin using bars
sns.countplot(x='Source', data=df1)

# Display the figure
plt.show()
```

## WHICH ARE THE MOST PROFITABLE Product, Payment Mode, Bank and Product amount?

```python
# Grouping products by Category and Sub-Category
profitable = pd.DataFrame(df1.groupby(['Product Code', 'Payment Status', 'Coupon code', 'Payment Mode', 'Currency  Code',
       'Transaction Bank']).sum()['Product Amount with GST'])

# Sorting the values
profitable.sort_values(['Product Code','Product Amount with GST'], ascending=False)
```
